```python
import statsmodels.api as sm
for response_var in ['ARS-총점', 'ARS-부주의', 'ARS-과잉', 'CBCL', '처리속도']:
    for explanatory_var in ['A4난이도_최대', 'A4결정시간_평균', 'A4결정시간_표준편차', 'A4터치_횟수', 'A5난이도_최대',
                  'A5결정시간_평균', 'A5결정시간_표준편차', 'A5터치_횟수', 'M5난이도_최대', 'M5남은_제한시간_평균',
                  'M5남은_제한시간_표준편차', 'M5남은_제한시간_최대', 'M5터치_횟수', 'V4왼쪽_정답율',
                  'V4오른쪽_정답율', 'V4난이도_최대', 'V4마지막_터치시간_평균', 'V4마지막_터치시간_표준편차',
                  'V4왼쪽_마지막_터치시간_평균', 'V4왼쪽_마지막_터치시간_표준편차', 'V4오른쪽_마지막_터치시간_평균',
                  'V4오른쪽_마지막_터치시간_표준편차', 'V4터치_횟수', 'V4왼쪽_터치_횟수', 'V4오른쪽_터치_횟수',
                  'V5난이도_최대', 'V5결정시간_평균', 'V5결정시간_표준편차', 'V5터치_횟수']:
        X = sb[explanatory_var]
        y = sb[response_var]

        # 상수항 추가
        X = sm.add_constant(X).

        # 회귀분석 모델 적합
        model = sm.OLS(y, X).fit()

        # 결과 출력
        print(f"반응변수: {response_var}, 설명변수: {explanatory_var}")
        print(model.summary())
        print("--------------------------------------------------")
```
