```python
import pandas as pd

t=pd.read_excel("C:/Users/user/Desktop/산프 우리소프트/분석자료/상관분석자료.xlsx")
t.to_csv(("C:/Users/user/Desktop/산프 우리소프트/분석자료/상관분석.csv"))
wt=pd.read_csv("C:/Users/user/Desktop/산프 우리소프트/분석자료/상관분석.csv",index_col=0)
wt.head()
```

```python
# 설명 변수 선택
explanatory_vars = ['A4정답율', 'A4난이도_평균', 'A4난이도_표준편차', 'A4난이도_최대', 'A4결정시간_평균', 'A4결정시간_표준편차',
                    'A4터치_횟수', 'A5정답율', 'A5난이도_평균', 'A5난이도_표준편차', 'A5난이도_최대', 'A5결정시간_평균',
                    'A5결정시간_표준편차', 'A5터치_횟수', 'M5정답율', 'M5난이도_평균', 'M5난이도_표준편차', 'M5난이도_최대',
                    'M5남은_제한시간_평균', 'M5남은_제한시간_표준편차', 'M5남은_제한시간_최대', 'M5터치_횟수', 'V4전체_정답율',
                    'V4난이도_평균', 'V4난이도_표준편차', 'V5정답율', 'V5난이도_평균', 'V5난이도_표준편차', 'V4왼쪽_정답율',
                    'V4오른쪽_정답율', 'V4난이도_최대', 'V4마지막_터치시간_평균', 'V4마지막_터치시간_표준편차',
                    'V4왼쪽_마지막_터치시간_평균', 'V4왼쪽_마지막_터치시간_표준편차', 'V4오른쪽_마지막_터치시간_평균',
                    'V4오른쪽_마지막_터치시간_표준편차', 'V4터치_횟수', 'V4왼쪽_터치_횟수', 'V4오른쪽_터치_횟수',
                    'V5난이도_최대', 'V5결정시간_평균', 'V5결정시간_표준편차', 'V5터치_횟수']


# 반응 변수 선택
response_vars = ['ARS-총점', 'ARS-부주의', 'ARS-과잉', 'CBCL', '처리속도']

```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 설명 변수와 반응 변수를 사용하여 회귀모델 학습
X = wt[explanatory_vars]
y = wt[response_vars]
model = sm.OLS(y, sm.add_constant(X))  # 회귀 모델 생성
results = model.fit()  # 모델 피팅

# 잔차 추출
residuals = results.resid

# 선형성 확인: 잔차 대비 예측값 산점도
plt.scatter(results.fittedvalues, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("예측값")
plt.ylabel("잔차")
plt.title("선형성: 잔차 대비 예측값 산점도")
plt.show()



```

```python
import statsmodels.api as sm
import scipy.stats as stats
import matplotlib.pyplot as plt

# 회귀모델을 통해 잔차 추출
residuals = results.resid

# 잔차의 히스토그램
plt.hist(residuals, bins=20, density=True, alpha=0.7)
plt.xlabel("잔차")
plt.ylabel("밀도")
plt.title("잔차의 분포")
plt.show()

# QQ 플롯 <-정규성 확인
sm.qqplot(residuals, line='s')
plt.xlabel("이론적인 분위수")
plt.ylabel("잔차의 분위수")
plt.title("QQ 플롯")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 잔차와 예측값 추출
residuals = results.resid
predicted = results.fittedvalues

# 잔차와 예측값의 산점도 그리기
plt.scatter(predicted, residuals, alpha=0.7)

# 가로선 추가: 잔차가 0 근처에서 등분산성을 나타내는 패턴을 보여줌
plt.axhline(y=0, color='r', linestyle='--')

# 축과 타이틀 설정
plt.xlabel("예측값")
plt.ylabel("잔차")
plt.title("등분산성: 잔차 대비 예측값 산점도")

# 그래프 출력
plt.show()
```

```python
import scipy.stats as stats

correlation_df = pd.DataFrame(columns=['Variable', 'Response Variable', 'Pearson Correlation', 'P-value'])

# 설명 변수와 반응 변수 간의 상관 관계 확인
for var in explanatory_vars:
    for resp_var in response_vars:
        corr, p_value = stats.pearsonr(wt[var], wt[resp_var])
        correlation_df = pd.concat([correlation_df, pd.DataFrame({'Variable': [var],
                                                              'Response Variable': resp_var,    
                                                              'Pearson Correlation': [corr],
                                                              'P-value': [p_value]})],
                               ignore_index=True)
correlation_df.to_csv("C:/Users/user/desktop/산프 우리소프트/분석자료/2023_07_18_상관.csv",index=False)
correlation_df
```

```python
# 설명 변수와 반응 변수 간의 상관 계수 계산
correlation_matrix = wt[explanatory_vars + response_vars].corr()
correlation_df = correlation_matrix.reset_index().rename(columns={'index': 'Variable'})

correlation_df.to_csv("C:/Users/user/Desktop/산프 우리소프트/분석자료/상관.csv", index=False)

correlation_df
```
